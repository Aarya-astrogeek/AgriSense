```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
retail = pd.read_csv("/content/retail_price_data (1).csv")
wholesale = pd.read_csv("/content/wholesale_price_data (2).csv")
```

```python
retail = retail.loc[:, ~retail.columns.str.contains("^Unnamed")]
wholesale = wholesale.loc[:, ~wholesale.columns.str.contains("^Unnamed")]
```

```python
retail = retail.rename(columns={
    "States/UTs": "state",
    "Date": "date"
})

wholesale = wholesale.rename(columns={
    "States/UTs": "state",
    "Date": "date"
})
```

```python
retail["date"] = pd.to_datetime(retail["date"], errors="coerce")
wholesale["date"] = pd.to_datetime(wholesale["date"], errors="coerce")
```

```python
retail_long = retail.melt(
    id_vars=["state", "date"],
    var_name="commodity",
    value_name="retail_price"
)
```

```python
wholesale_long = wholesale.melt(
    id_vars=["state", "date"],
    var_name="commodity",
    value_name="wholesale_price"
)
```

```python
retail_long["retail_price"] = pd.to_numeric(
    retail_long["retail_price"], errors="coerce"
)

wholesale_long["wholesale_price"] = pd.to_numeric(
    wholesale_long["wholesale_price"], errors="coerce"
)
```

```python
bucket4_crops = [
    "Gram Dal",
    "Gur",
    "Groundnut Oil (Packed)",
    "Urad Dal",
    "Tur/Arhar Dal",
    "Soya Oil (Packed)"
]

retail_long = retail_long[retail_long["commodity"].isin(bucket4_crops)]
wholesale_long = wholesale_long[wholesale_long["commodity"].isin(bucket4_crops)]
```

```python
retail_long["retail_price"] = retail_long.groupby(
    ["state", "commodity"]
)["retail_price"].transform(lambda x: x.fillna(x.median()))

wholesale_long["wholesale_price"] = wholesale_long.groupby(
    ["state", "commodity"]
)["wholesale_price"].transform(lambda x: x.fillna(x.median()))
```

```python
merged = pd.merge(
    retail_long,
    wholesale_long,
    on=["state", "date", "commodity"],
    how="inner"
)
```

```python
merged["spread"] = merged["retail_price"] - merged["wholesale_price"]
merged["spread_pct"] = merged["spread"] / merged["wholesale_price"]
```

```python
def prepare_series(df, state, crop, column="wholesale_price"):
    ts = df[
        (df["state"] == state) &
        (df["commodity"] == crop)
    ].sort_values("date")

    ts = ts.set_index("date")[column]
    ts = ts.asfreq("D")
    ts = ts.fillna(method="ffill")
    return ts
```

```python
def prepare_series(df, state, crop, column="wholesale_price"):
    ts = df[
        (df["state"] == state) &
        (df["commodity"] == crop)
    ].sort_values("date")

    ts = ts.set_index("date")[column]
    ts = ts.asfreq("D")
    ts = ts.fillna(method="ffill")
    return ts
```

```python
def train_test_split(ts, test_days=30):
    train = ts[:-test_days]
    test = ts[-test_days:]
    return train, test
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
def run_arima(train, test):
    model = ARIMA(train, order=(1,1,1))
    fit = model.fit()
    forecast = fit.forecast(steps=len(test))
    rmse = np.sqrt(mean_squared_error(test, forecast))
    return rmse, forecast
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
def run_sarima(train, test):
    model = SARIMAX(
        train,
        order=(1,1,1),
        seasonal_order=(1,0,1,12),  # monthly seasonality
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    fit = model.fit(disp=False)
    forecast = fit.forecast(steps=len(test))
    rmse = np.sqrt(mean_squared_error(test, forecast))
    return rmse, forecast
```

```python
from prophet import Prophet
```

```python
def run_prophet(train, test):
    df_p = train.reset_index()
    df_p.columns = ["ds", "y"]

    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False
    )
    model.fit(df_p)

    future = model.make_future_dataframe(periods=len(test))
    forecast = model.predict(future)

    preds = forecast.tail(len(test))["yhat"].values
    rmse = np.sqrt(mean_squared_error(test.values, preds))
    return rmse, preds
```

```python
def compare_models(ts):
    train, test = train_test_split(ts)

    results = {}

    rmse_a, _ = run_arima(train, test)
    results["ARIMA"] = rmse_a

    rmse_s, _ = run_sarima(train, test)
    results["SARIMA"] = rmse_s

    rmse_p, _ = run_prophet(train, test)
    results["Prophet"] = rmse_p

    best_model = min(results, key=results.get)
    return results, best_model
```

```python
bucket4_results = []

for (state, crop), _ in merged.groupby(["state", "commodity"]):
    try:
        ts = prepare_series(merged, state, crop)

        if len(ts) < 120:
            continue

        scores, best = compare_models(ts)

        bucket4_results.append({
            "state": state,
            "commodity": crop,
            "ARIMA_RMSE": scores["ARIMA"],
            "SARIMA_RMSE": scores["SARIMA"],
            "Prophet_RMSE": scores["Prophet"],
            "Best_Model": best
        })

    except Exception as e:
        print(state, crop, e)
```

```python
def prepare_wholesale_series(df, state, crop):
    ts = (
        df[(df["state"] == state) & (df["commodity"] == crop)]
        .sort_values("date")
        .set_index("date")["wholesale_price"]
        .astype(float)
    )

    ts = ts.asfreq("D")
    ts = ts.interpolate()

    return ts
```

```python
def train_test_split(ts, horizon=30):
    return ts[:-horizon], ts[-horizon:]
```

```python
import numpy as np

def avg_price_deviation(actual, predicted):
    return np.mean(np.abs(np.array(actual) - np.array(predicted)))
```

```python
def naive_forecast(train, steps):
    return np.repeat(train.iloc[-1], steps)
```

```python
def moving_average_forecast(train, steps, window):
    avg = train.rolling(window).mean().iloc[-1]
    return np.repeat(avg, steps)
```

```python
from statsmodels.tsa.holtwinters import Holt

def holt_forecast(train, steps):
    model = Holt(train)
    fitted = model.fit(optimized=True)
    return fitted.forecast(steps)
```

```python
ts = prepare_wholesale_series(
    merged,
    state="Maharashtra",
    crop="Gram Dal"
)

train, test = train_test_split(ts, horizon=30)

# Naive
naive_pred = naive_forecast(train, len(test))
naive_err = avg_price_deviation(test.values, naive_pred)

# MA 7
ma7_pred = moving_average_forecast(train, len(test), window=7)
ma7_err = avg_price_deviation(test.values, ma7_pred)

# MA 14
ma14_pred = moving_average_forecast(train, len(test), window=14)
ma14_err = avg_price_deviation(test.values, ma14_pred)

# Holt
holt_pred = holt_forecast(train, len(test))
holt_err = avg_price_deviation(test.values, holt_pred)

naive_err, ma7_err, ma14_err, holt_err
```

```python
import numpy as np

alpha_values = np.linspace(0.1, 0.9, 9)
beta_values = np.linspace(0.1, 0.9, 9)
```

```python
from statsmodels.tsa.holtwinters import Holt

def holt_grid_search(train, test):
    best_error = float("inf")
    best_params = None

    for alpha in alpha_values:
        for beta in beta_values:
            try:
                model = Holt(train)
                fitted = model.fit(
                    smoothing_level=alpha,
                    smoothing_trend=beta,
                    optimized=False
                )

                forecast = fitted.forecast(len(test))
                error = avg_price_deviation(test.values, forecast.values)

                if error < best_error:
                    best_error = error
                    best_params = (alpha, beta)

            except:
                continue

    return best_error, best_params
```

```python
best_error, best_params = holt_grid_search(train, test)

best_error, best_params
```

```python
alpha, beta = best_params

final_model = Holt(train).fit(
    smoothing_level=alpha,
    smoothing_trend=beta,
    optimized=False
)

final_forecast = final_model.forecast(len(test))

avg_price_deviation(test.values, final_forecast.values)
```
